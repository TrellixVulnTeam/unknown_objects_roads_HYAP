```python
import numpy as np
import open3d as o3d
from laserscan import SemLaserScan
import yaml
```

```python
from bounding_box import BoundingBox
```

```python
with open("kitti/semantic-kitti-all.yaml", "r") as stream:
    try:
        CFG = yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)
```

```python
scan = SemLaserScan(sem_color_dict=CFG['color_map'])
scan.open_scan("kitti/velodyne/000000.bin")
scan.open_label("kitti/velodyne/000000_pred.label")
scan.colorize()
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(scan.points)
#pcd.paint_uniform_color([0.6,0.6,0.6])
#pcd.colors = o3d.utility.Vector3dVector(scan.sem_label_color)
pcd.paint_uniform_color([0.6,0.6,0.6])
```

```python
o3d.visualization.draw_geometries([pcd])
```
